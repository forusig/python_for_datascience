# Functional

```python
def nilai(df, index=0, count=0, max_count=10):
    # Cek apakah count kurang dari max_count
    if count < max_count:
        # Cek apakah index lebih kecil dari jumlah baris DataFrame
        if index < len(df):
            # Ambil nilai Exam_Score, Hours_Studied, dan Parental_Involvement pada index saat ini
            exam_score = df.loc[index, 'Exam_Score']
            hours_studied = df.loc[index, 'Hours_Studied']
            parental_involvement = df.loc[index, 'Parental_Involvement']

            # Jika nilai Exam_Score lebih dari 75, cetak
            if exam_score > 75:
                print(f'Index {index}, Exam Score: {exam_score}, Hours Studied: {hours_studied}, Parental Involvement: {parental_involvement}')
                # Tambah count
                count += 1

            # Panggil fungsi ini lagi untuk index berikutnya
            nilai(df, index + 1, count, max_count)

# Contoh pemanggilan fungsi
nilai(df)
```

```python
# Fungsi untuk memfilter DataFrame menggunakan fungsi filter
def filter(df, func):
    return df[df.apply(func, axis=1)]

# Fungsi filter untuk memeriksa nilai ujian di atas 75
def score(row):
    return row['Exam_Score'] > 75

# Fungsi filter untuk memeriksa jam belajar di atas 5
def hours(row):
    return row['Hours_Studied'] > 5

# Menggabungkan kedua kondisi
def score_and_hours(row):
    return score(row) and hours(row)

# Menggunakan filter dengan fungsi filter yang telah ditentukan
high_scores = filter(df, score)
high_hours = filter(df, hours)
combined_filter = filter(df, score_and_hours)

print("Jumlah siswa dengan Exam Score lebih dari 75:", len(high_scores))
print("Jumlah siswa dengan Hours Studied lebih dari 5:", len(high_hours))
print("Jumlah siswa dengan Exam Score lebih dari 75 dan Hours Studied lebih dari 5:", len(combined_filter))
```

# Numpy

```python
import numpy as np

exam_scores = df['Exam_Score'].values  # Mengambil nilai Exam_Score sebagai array NumPy
mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)
std_deviation = np.std(exam_scores)

print("Statistik Exam Score:")
print(f"Rata-rata: {mean_score}")
print(f"Median: {median_score}")
print(f"Standar Deviasi: {std_deviation}")
```

```python
Hours_Studied_array = np.array(df['Hours_Studied'])
exam_scores_array = np.array(df['Exam_Score'])

join = np.column_stack((Hours_Studied_array, exam_scores_array))
print("\nData setelah penggabungan")
print(join [:10])
```

```python
splits = np.array_split(exam_scores, 2)  # Membagi Exam Score menjadi 2 bagian
print("\nHasil pemisahan Exam Scores:")
for i, split in enumerate(splits):
    print(f"Bagian {i + 1}: {split}")
```

```python
# Menghitung matriks korelasi antara Hours_Studied dan Exam_Score
matrix_data = df[['Hours_Studied', 'Exam_Score']].values
correlation_matrix = np.corrcoef(matrix_data, rowvar=False)

print("\nMatriks Korelasi:")
print(correlation_matrix)
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membuat histogram untuk distribusi Exam_Score
plt.figure(figsize=(8, 5))
plt.hist(df['Exam_Score'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribusi Exam_Score')
plt.xlabel('Exam_Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Membuat scatter plot antara Hours_Studied dan Exam_Score
plt.figure(figsize=(8, 5))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='purple')
plt.title('Scatter Plot antara Hours_Studied dan Exam_Score')
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.grid(True)
plt.show()
```

```python
# Scatterplot dengan Seaborn, dikategorikan berdasarkan Family_Income
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', hue='Family_Income', data=df, palette='deep', s=100)
plt.title('Scatter Plot Hours_Studied vs Exam_Score berdasarkan Family_Income')
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.show()
```

```python
# Hanya mengambil kolom yang numerik
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Membuat heatmap dari korelasi antar kolom numerik
plt.figure(figsize=(8, 5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```
